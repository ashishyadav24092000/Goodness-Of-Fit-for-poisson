## Chi-square Goodness Of Fit test::

```python
import scipy
from scipy.stats import chi2
from scipy.stats  import poisson
```

```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_excel("/home/ashish/projects/GoodnessOfFit_PoissonDistribution/P_distribution.xlsx")
dataframe
```

```python
observed_freq = dataframe['Frequency']
observed_freq
```

```python
total_arrival = 600
total_time_period  = 100
mu = total_arrival/total_time_period
```

```python
mu
```

```python
Expected_frequency = []
for i in range(len(observed_freq)):
    E_freq = 100*poisson.pmf(i,mu)
    Expected_frequency.append(E_freq)
```

```python
Expected_frequency
```

```python
Expected_frequency_new = [round(element,2) for element in Expected_frequency]
Expected_frequency_new
```

```python
df = pd.DataFrame(list(zip(observed_freq,Expected_frequency_new)),columns=['Observed Frequency','Expected Frequency'])
df
```

```python
observed_frequency = list(df['Observed Frequency'])
expected_frequency = list(df['Expected Frequency'])
```
